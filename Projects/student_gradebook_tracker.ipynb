### üìò **Student Gradebook Tracker**

**Tech Stack**: Python, Pandas, Matplotlib
**Description**:
Built a Gradebook Tracker to manage and analyze student performance efficiently. The system allows users to input student marks for multiple subjects, calculates individual averages, assigns ranks, and visualizes grade distribution using bar and pie charts.

**Features**:

* ‚úÖ Input system for multiple students and subjects
* üìä Computes average marks and percentage per student
* üèÖ Ranks students based on overall performance
* üìö Assigns grades based on score ranges (A, B, C, etc.)
* üìà Visualizes:

  * Subject-wise average scores (Bar Chart)
  * Grade distribution (Pie Chart)

**Skills Demonstrated**:

* üß† Pandas DataFrame operations (creation, filtering, sorting, merging)
* üîÅ `groupby()` and `apply()` functions for aggregation
* üéØ Data cleaning and formatting
* üìâ Visualization using Matplotlib (custom bar and pie charts)
* üêç Python scripting for CLI interaction

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def assign_grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 80:
        return 'B'
    elif avg >= 70:
        return 'C'
    elif avg >= 60:
        return 'D'
    else:
        return 'F'


with open('data.json', 'r') as file:
    data = json.load(file)

df = pd.DataFrame(data)
df['Total'] = df[['Math','Science','English']].sum(axis=1)
df['Average'] = (df['Total']/3).astype(int)
df['Grade'] = df['Average'].apply(assign_grade) 
df['Rank'] = df['Total'].rank(ascending=False,method='min').astype(int)

print("üìã Final Gradebook:\n")
df.sort_values(by='Rank')
```

```python
subject_avg = df[['Math', 'Science', 'English']].mean()
subject_avg.plot(kind='bar', title='Subject-wise Average Marks', color='skyblue')
plt.ylabel('Average Marks')
plt.tight_layout()
```

```python
grade_dist = df['Grade'].value_counts()
grade_dist.plot(kind='pie', autopct='%1.1f%%', title='Grade Distribution', colors=['#4CAF50','#FFC107','#2196F3','#FF5722','#9E9E9E'])
plt.ylabel('')
plt.tight_layout()
```
