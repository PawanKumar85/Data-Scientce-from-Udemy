## Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x**3
```

```python
plt.plot(x,y)
plt.title("Matplotlib")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
## Sub plot's
plt.subplot(2,2,1)
plt.plot(x**2,x**2)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
# Object Oriented Way
fig = plt.figure()
axis1 = fig.add_axes([0.5,0.5,0.5,0.5])  # Left,Bottom,Hight,Width
axis1.plot(x,y)
axis2 = fig.add_axes([1,0.5,0.5,0.5])  # Left,Bottom,Hight,Width
axis2.plot(x,y) 
```

```python
# Types of plot
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```

```python
# SAving Plots
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
plt.savefig("matplotlib1.png")
```

```python
# Working with images
import matplotlib.image as mpimg

img = mpimg.imread("matplotlib1.png")
plt.imshow(img)
plt.axis("off")
```
