# Seaborn

## Distribution Plot
- We use distribution plots to visualize the distribution of quantitative data.
* distplot/histplot
* jointplot
* pairplot
* rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ds = sns.load_dataset("tips")
ds
```

```python
plt.subplot(1,2,1)
sns.histplot(ds["total_bill"],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(ds["tip"],kde=True)
```

```python
# Jointplot
sns.jointplot(x="total_bill",y="tip",data=ds)
```

```python
sns.jointplot(x="total_bill",y="tip",data=ds,kind='hist')
```

```python
# Pairplot
sns.pairplot(ds,hue='size',palette="rainbow")
```

```python
# rugplot

sns.rugplot(ds)
```

# Categorical Data Plots
- boxplot
- violinplot
- stripplot
- swarmplot
- barplot
- countplot

```python
sns.countplot(x = ds['sex'],hue=ds["smoker"])
```

```python
sns.barplot(x=ds["sex"],y=ds["total_bill"])
```

```python
# boxplot
sns.boxplot(x="tip",y="day",data=ds,palette='rainbow')
```

```python
# violinplot
sns.violinplot(x = 'day',y="total_bill",data=ds,palette="rainbow")
```
