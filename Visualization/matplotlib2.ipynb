```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Plot Line Graph

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x,y,color="red",linestyle=":",marker="s")
plt.title("Simple Plot Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

## Plot Bar Graph

```python
categories = ['Python', 'Java', 'C++', 'JavaScript']
popularity = [85, 75, 60, 70]

plt.bar(categories,popularity,color='skyblue')
plt.title("Programming Language Popularity")
plt.xlabel("Languages")
plt.ylabel("Popularity %")
```

## Plot Pie Chart

```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

plt.pie(sizes,labels=labels,autopct="%1.1f%%",startangle=90)  # adds names,shows percentages,rotates start point,
plt.title("Programming Language Market Share") # ensures it's a perfect circle
plt.axis("equal")
```

## Plot Histograms Chart

```python
ages = [18, 19, 21, 21, 22, 23, 24, 24, 24, 25, 26, 27, 30, 30, 35]
plt.hist(ages,bins=5,color="orange",edgecolor="black") 
## bins=5 means divide age range into 5 buckets
## edgecolor='black' gives nice borders around bars
plt.title("Age Distribution")
plt.xlabel("Age Groups")
plt.ylabel("Frequency")
```

## Subplots & Multiple Charts

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [30, 25, 20, 10]

plt.subplot(1, 2, 1)
plt.plot(x,y1,color="green")
plt.title("Line 1")

plt.subplot(1,2,2)
plt.plot(x,y2,color="red")
plt.title("Line 2")

plt.tight_layout()
```

## Legends and Grid

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [30, 25, 20, 10]

plt.plot(x,y1,color="green",label='Ascending')
plt.plot(x,y2,color="red",label='Descending')

plt.title("Legend and Grid Example")  # label='text' assigns label to line
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.legend() # plt.legend() shows the legend box.
plt.grid(True)  # plt.grid(True) enables grid lines.
```

# Intermediate

```python
fig,axs = plt.subplots(2, 2,figsize=(8, 6))

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

axs[0,0].plot(x,y,color="green")
axs[0,0].set_title("Plot 1")

axs[0, 1].bar(x, y, color='blue')
axs[0, 1].set_title("Plot 2")

axs[1, 0].scatter(x, y, color='red')
axs[1, 0].set_title("Plot 3")

axs[1, 1].pie(y, labels=x)
axs[1, 1].set_title("Plot 4")

plt.tight_layout()
```

- `fig, axs = plt.subplots(rows, cols)` returns a figure and axes array
- `axs[row, col]` allows you to access each plot
- `figsize=(width, height)` changes figure size in inches

# Advanced Level

```python
x = [1, 2, 3, 4]
temp = [30, 32, 34, 36]
humidity = [80, 70, 60, 50]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x,y,'g-',label="Temperature")
ax2.plot(x, humidity, 'b--', label="Humidity")

ax1.set_xlabel("Day")
ax1.set_ylabel("Temperature (°C)", color='g')
ax2.set_ylabel("Humidity (%)", color='b')

plt.title("Twin Axes Example")
```

# Expert Matplotlib Tools

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title("3D Surface Plot")
```

### Animations (Intro)

```python
import matplotlib.animation as animation

fig, ax = plt.subplots()
x, y = [], []
line, = ax.plot([], [], 'r-')

def update(frame):
    x.append(frame)
    y.append(frame ** 0.5)
    line.set_data(x, y)
    ax.relim()
    ax.autoscale_view()
    return line,

ani = animation.FuncAnimation(fig, update, frames=range(100), blit=True)
```
