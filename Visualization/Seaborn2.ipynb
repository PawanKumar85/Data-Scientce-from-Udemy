# Seaborn
---
**Seaborn is built on top of Matplotlib and makes statistical plots beautiful and super easy.**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
# Scatter Plot
sns.scatterplot(data=tips,x="total_bill",y="tip")
plt.title("Total Bill vs Tips")
```

```python
# barplot
sns.barplot(data=tips,x="total_bill",y="day",hue='sex')
plt.title("Average Bill by Day & Gender")
```

```python
# sns.lineplot
sns.lineplot(data=tips,x='size',y='tip')
plt.title("Tip Amount vs Party Size")
```

```python
# Catplot
sns.catplot(data=tips,x='day',y='total_bill',col='sex')
```

```python
sns.catplot(data=tips, x="day", y="total_bill", row="time", kind="swarm")
```

```python
sns.pairplot(data=tips,hue="sex")
```

```python
# heatmap() 
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
plt.title("Total Bill Distribution per Day")
```

```python
sns.set_style("whitegrid")
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
```

```python
sns.pairplot(data=sns.load_dataset("iris"), hue="species")
```
